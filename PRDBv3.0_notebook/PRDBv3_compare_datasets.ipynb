### Venn diagram for 3-way intersection : PRDBv2.0

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn3, venn3_circles
plt.rcParams['font.family'] = "Times New Roman"

# PDB IDs between the three groups. PRDBv2, Juan, and Zou datasets
# Convert the 'ID' columns to sets
set1 = set(nithinChandran_126_df['PDB']) 
set2 = set(parezCano_106_df['PDB'])
set3 = set(huangSheng_72_df['PDB'])

# Create a Venn diagram
plt.figure(layout='constrained',figsize=(8, 6))
#venn = venn3([set1, set2, set3], set_labels=None, alpha=0.7, )
venn = venn3([set1, set2, set3], ('Nithin', 'Parez-Cano', 'Huang (Zou)'), alpha=0.7, )
circ = venn3_circles([set1, set2, set3], linestyle='solid', linewidth=1, color="grey")

# Customizing the plot
# Add titles
#plt.title("Venn Diagram of PDB IDs Across Three DataFrames", fontsize=16, fontweight='bold', pad=20)

# Customize circle colors and font sizes
for text in venn.set_labels:
    text.set_fontsize(14)
    text.set_fontweight('bold')

# Zou group's count location
txt_zou = venn.get_label_by_id('011')
x, y = txt_zou.get_position()
txt_zou.set_position((x, y + -0.1))

# Add labels to the circles
for text in venn.subset_labels:
    if text:  # Avoid None values
        text.set_fontsize(14)

# Add gridlines for better alignment
plt.grid(visible=False)

# show the plot
plt.show()
```

#### UpSet plot: to find intersection between PRDBv3, PRDBv2, Juan and Zou docking benchmarks

```python
# This code is used to generate the UpSet plot of the PRDBv3 comprehensive benchmark dataset.
# Import necessary libraries
import pandas as pd
from upsetplot import UpSet, generate_counts, plot
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = "Times New Roman"

# Convert the 'ID' columns to sets
set1 = set(PRDBv3_197_df['PDB']) 
set2 = set(nithinChandran_126_df['PDB'])
set3 = set(parezCano_106_df['PDB'])
set4 = set(huangSheng_72_df['PDB'])

# Create a DataFrame to hold the intersection information
# Combine sets into a DataFrame
set_names = ['PRDBv3', 'Nithin et al.', 'Parez-Cano et al.', 'Huang et al.']
all_elems = set1.union(set2).union(set3).union(set4)
df = pd.DataFrame([[e in set1, e in set2, e in set3, e in set4] for e in all_elems], columns = set_names)
df_up = df.groupby(set_names).size()

# Define figure size and style
fig = plt.figure(figsize=(12, 6))
plt.rcParams["font.size"] = 14

# Create an UpSet plot
upset = UpSet(df_up, orientation='horizontal', show_counts='%d', sort_by='cardinality', min_subset_size=1, facecolor="darkblue",
              shading_color="lightgray", other_dots_color=0.4, element_size=None, intersection_plot_elements=10, totals_plot_elements=2)

# Customize the plot
upset.style_categories('PRDBv3', shading_edgecolor="darkorange", shading_linewidth=1)
upset.style_subsets(min_degree=1, facecolor="firebrick", edgecolor="black", linewidth=0.5)
upset.style_subsets(min_degree=2, facecolor="darkgreen", edgecolor="black", linewidth=.05)
upset.style_subsets(min_degree=3, facecolor="gold", edgecolor="black", linewidth=0.5)
upset.style_subsets(min_degree=4, facecolor="darkcyan", edgecolor="black", linewidth=0.5)

# Add labels to the plot
plot_result = upset.plot()
#plt.title("UpSet Plot of Protein-RNA Docking Benchmarks", fontsize=14, fontweight='bold')

# Update plot labels with adjusted font sizes
plot_result['intersections'].set_ylabel('Number of Structures', fontsize=14)
#plot_result['totals'].set_xlabel('Number of Structures', fontsize=10)

# Set the face color of the intersections
#plot_result['intersections'].set_facecolor('lightblue')

# Add a title to the plot
#plt.suptitle('UpSet Plot of Set Intersections', fontsize=10, fontweight='bold', y=1.05)
plt.subplots_adjust(left=0.1, right=1 ,top=0.85, bottom=0.20)
# Save figure
#plt.savefig('/home/shrikant/Projects/git_dev//PRDBv3_dataset/figures/Figure_3.tiff', format='tiff', dpi=300, bbox_inches='tight')
#plt.savefig('/home/shrikant/Projects/git_dev/PRDBv3_dataset/figures/Figure_3.png', format='png', dpi=600, bbox_inches='tight')

# Show the plot
plt.show()
plt.close()
```

#### Number of structures (% similarity with PRDBv3)

```python
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['font.family'] = "Times New Roman"
plt.rcParams["font.size"] = 14  # Increased font size for readability

# Data
bardata = {'Nithin et al.': 126, 'Parez-Cano et al.': 51, 'Huang et al.': 36}

# Total for calculating percentages
total = sum(bardata.values())

# Create the bar chart using matplotlib
plt.figure(figsize=(4, 6))  # Slightly larger size for better spacing

# Plot bars
bars = plt.bar(bardata.keys(), bardata.values(), color='darkblue', edgecolor='black', linewidth=0.8, alpha=0.85, width=0.3)

# Annotate the bars with counts and percentages
for i, (key, value) in enumerate(bardata.items()):
    percentage = (value / total) * 100
    # Position annotation slightly above the bars, avoid overlap
    plt.text(i, value + 2, f"{value} ({percentage:.1f}%)", color='black', ha='center', fontsize=14)

# Add labels and title with improved font size
plt.xlabel("Benchmark Datasets", fontsize=18)
plt.ylabel("Number of Structures\n(Percentage Similarity with PRDBv3)", fontsize=18)
plt.title("Comparison of Benchmark Datasets", fontsize=18, fontweight='bold')

# Improve xticks and yticks readability
#plt.xticks(x=None)
plt.yticks(np.arange(0, 140+1, 20), fontsize=16)

# Add horizontal grid lines for better readability
plt.grid(axis='y', linestyle='-', alpha=0.6)

# Remove the top and right spines for a cleaner look
for spine in plt.gca().spines.values():
    if spine.spine_type != 'bottom' and spine.spine_type != 'left':
        spine.set_visible(False)

# Improve spacing between elements to avoid crowding
plt.subplots_adjust(top=0.85, bottom=0.15)
#plt.savefig('/home/shrikant/Projects/git_dev/PRDBv3_dataset/figures/PRDBv3_benchmark_bar_notext.png', format='png', dpi=600, bbox_inches='tight')
# Show plot
plt.show()
```
